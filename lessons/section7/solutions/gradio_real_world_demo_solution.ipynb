# Section 7: Gradio Real World Demo — Solution

This notebook contains the solution for the Image Brightness Adjuster exercise from `gradio_real_world_demo.ipynb`.

## Solution: Image Brightness Adjuster

```python
import gradio as gr
from PIL import Image, ImageEnhance

# 1. Define the function
def adjust_brightness(image, brightness):
    enhancer = ImageEnhance.Brightness(image)
    return enhancer.enhance(brightness)

# 2. Create the Gradio Interface
exercise_demo = gr.Interface(
    fn=adjust_brightness,
    inputs=[
        gr.Image(type="pil", label="Upload an Image"),
        gr.Slider(minimum=0.0, maximum=2.0, step=0.1, value=1.0, label="Brightness")
    ],
    outputs=gr.Image(label="Result"),
    title="Image Brightness Adjuster",
    description="Upload an image and adjust the brightness using the slider."
)

# 3. Launch the demo
exercise_demo.launch()
```
