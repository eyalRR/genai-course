## Full Example: Image Filter App

This app lets a user upload any image and apply one of three filters:

- **Grayscale** — removes all color
- **Blur** — softens the image
- **Edge Detection** — highlights edges and outlines

The key new Gradio components here are:
- `gr.Image(type="pil")` — accepts an uploaded image and passes it directly as a PIL Image object
- `gr.Dropdown()` — lets the user pick from a list of options
- `gr.Image()` as output — displays the processed image

````python
import gradio as gr
from PIL import Image, ImageFilter

def apply_filter(image, filter_name):
    if filter_name == "Grayscale":
        return image.convert("L").convert("RGB")  # convert back to RGB so Gradio displays it correctly
    elif filter_name == "Blur":
        return image.filter(ImageFilter.GaussianBlur(radius=5))
    elif filter_name == "Edge Detection":
        return image.filter(ImageFilter.FIND_EDGES)

demo = gr.Interface(
    fn=apply_filter,
    inputs=[
        gr.Image(type="pil", label="Upload an Image"),
        gr.Dropdown(choices=["Grayscale", "Blur", "Edge Detection"], label="Choose a Filter", value="Grayscale")
    ],
    outputs=gr.Image(label="Filtered Image"),
    title="Image Filter App",
    description="Upload any image and apply a filter to see the result instantly."
)

demo.launch()
````

**What's happening:**

1. `gr.Image(type="pil")` — the `type="pil"` argument tells Gradio to pass the uploaded image directly as a `PIL.Image` object to your function, so you can work with it immediately.
2. `gr.Dropdown(choices=[...], value="Grayscale")` — creates a dropdown menu. `value` sets the default selected option.
3. The function receives both inputs, applies the chosen filter using PIL, and returns the processed image.
4. `gr.Image()` as output renders the returned PIL image directly in the browser.
