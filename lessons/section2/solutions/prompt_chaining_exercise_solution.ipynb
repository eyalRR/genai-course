# Paris Trip Planning with Prompt Chaining

## Overview
Prompt chaining is a technique used to break down complex tasks into sequential subtasks, where the output of one task becomes the input for the next.

## Exercise Objective
Create a prompt chain to plan a comprehensive Paris trip using two subtasks.

### Subtask 1: Identify Key Attractions
**Prompt:** "What are the main tourist attractions in Paris?"

### Subtask 2: Create Detailed Itinerary
**Prompt:** "Based on the list of attractions you provided, suggest a detailed 3-day itinerary for visiting Paris."

## Python Implementation

```python
from openai import OpenAI

client = OpenAI()

# Define tasks for Paris trip planning
tasks = [
	"What are the main tourist attractions in Paris?",
	"Based on the list of attractions you provided, suggest a detailed 3-day itinerary for visiting Paris."
]

responses = []

# First task: Get list of attractions
first_response = client.chat.completions.create(
	model="gpt-4o-mini",
	messages=[
		{"role": "developer", "content": "You are a travel expert specializing in Paris tourism."},
		{"role": "user", "content": tasks[0]}
	]
)
responses.append(first_response.choices[0].message.content)

# Second task: Create itinerary based on attractions
second_response = client.chat.completions.create(
	model="gpt-4o-mini",
	messages=[
		{"role": "developer", "content": "You are a travel expert specializing in Paris tourism."},
		{"role": "user", "content": f"Previous attractions list: {responses[0]}"},
		{"role": "user", "content": tasks[1]}
	]
)
responses.append(second_response.choices[0].message.content)

# Print the responses to see the progression
for i, response in enumerate(responses, 1):
	print(f"\n--- Task {i} Response ---")
	print(response)
```
